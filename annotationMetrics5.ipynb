```python
def get(gt, column_index):
    indexes = []
    for index, row in gt.iterrows():
        if row.iloc[column_index] == 1:
            indexes.append(index)
    return indexes
```

```python
def clean_columns(df):
  return df.drop(df.columns[0:17], axis=1)
```

```python
def swap_columns(df, a, b):
  col1, col2 = a, b
  cols = list(df.columns)
  idx1, idx2 = cols.index(a), cols.index(b)
  cols[idx1], cols[idx2] = cols[idx2], cols[idx1]
  return df[cols]
```

```python
def print_cols(df):
  cols = list(df.columns)
  print(cols)
  return
```

```python
def ground_truth_calculation(df, size):
  size = (int)(size / 2 + 0.5)
  for index, row in df.iterrows():
    for column in df.columns:
        if row[column] >= size:
            df.loc[index, column] = 1
        else:
            df.loc[index, column] = 0
  return df
```

```python
def dfs_sum(dfs):
  df_sum = pd.concat(dfs).groupby(level=0).sum()
  groundTruth = df_sum.copy()
  groundTruth = ground_truth_calculation(groundTruth, len(dfs))
  return groundTruth.sum(axis=0)
```

```python
def dfs_sum2(dfs):
  df_sum = pd.concat(dfs).groupby(level=0).sum()
  groundTruth = df_sum.copy()
  groundTruth = ground_truth_calculation(groundTruth, len(dfs))
  return groundTruth
```

```python
def change_columns(dfs, newColumns):
  for df in dfs:
    df.columns = newColumns
  return
```

```python
def change_names(dfs, add):
  columns = list(dfs[0].columns)
  for i in range(len(columns)):
    columns[i] = columns[i] + '_' + add
  for df in dfs:
    df.columns = columns
```

```python
def merge(dfs, boolean):
    merged = pd.DataFrame()
    iter = 0
    for df in dfs:
        iter = iter + 1
        if (boolean) and iter == 2:
          df['dummy_column'] = df.index-1
        else:
          df['dummy_column'] = df.index
    for df in dfs:
        if merged.empty:
            merged = df
        else:
            merged = pd.merge(merged, df, on='dummy_column')
    merged.drop(columns='dummy_column', inplace=True)
    return merged
```

```python
def calculate_cronbach_alpha(df):
    alpha_results = {}
    for category in range(1, 22):
        columns = []
        for column in list(df.columns):
            cate = column.split('_')[0]
            if (int)(cate) == category:
                columns.append(column)
        temp = df[columns]
        alpha = reliability.cronbach_alpha(temp)
        alpha_results[category] = alpha[0]
    return alpha_results
```

```python
def print_results(results):
    valid = []
    for category, alpha in results.items():
        if (alpha > 0.7):
            valid.append(category)
            #print(f"Cronbach's alpha for {category}: {alpha}")
    #print(valid)
    return valid
```

```python
def list_of(results):
    list = []
    for category, alpha in results.items():
        if math.isnan(alpha):
            alpha = 0
        list.append(round(alpha,5))
    return list
```
